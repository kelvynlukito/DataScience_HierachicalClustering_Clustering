# **Hierarchical Clustering**

# **Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Import Dataset**

```python
df = pd.read_csv('/content/Mall_Customers.csv')
df
```

```python
X = df.iloc[:,[3,4]].values
print(X)
```

# **Using Dendrogram to Find Optimal Clusters**

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

# **Training Hierachical Clustering Model on Dataset**

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc_5 = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
hc_3 = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
```

```python
y_hc_5 = hc_5.fit_predict(X)
y_hc_3 = hc_3.fit_predict(X)
```

# **Visualising Cluster**

```python
for i in range(len(np.unique(y_hc_5))):
  plt.scatter(X[y_hc_5==i, 0] ,X[y_hc_5 == i, 1], label=f'Cluster {i+1}')
  plt.legend()
  plt.title('HC Cluster 5')
  plt.xlabel('Customer Income')
  plt.ylabel('Spending Score')
```

```python
for i in range(len(np.unique(y_hc_3))):
  plt.scatter(X[y_hc_3==i, 0] ,X[y_hc_3 == i, 1], label=f'Cluster {i+1}')
  plt.legend()
  plt.title('HC Cluster 3')
  plt.xlabel('Customer Income')
  plt.ylabel('Spending Score')
```
